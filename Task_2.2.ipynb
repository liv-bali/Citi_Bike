## Import and concatenate

```python
import pandas as pd 
import numpy as np
import os 
import requests
import json
from datetime import datetime
```

```python
# Create a list with all files in the folder using a list compehension

folderpath = r"Data" # make sure to put the 'r' in front
filepaths  = [os.path.join(folderpath, name) for name in os.listdir(folderpath)]
```

```python
filepaths
```

```python
# Read and concatenate all files simultaneously

df = pd.concat((pd.read_csv(f) for f in filepaths), ignore_index = True)
```

```python
## We start by defining the folder path where the Citi Bike CSV files are stored and use a list comprehension to create a list of file paths. 
# This allows us to automatically gather all CSV files in the directory without manually listing them. 
# Then, we use pd.concat() to read and merge all the files at once efficiently. The loop pd.read_csv(f) for f in filepaths reads each file into a 
# DataFrame, and pd.concat() combine them into a single dataset. Setting ignore_index=True ensures the index is reset, 
# creating a smooth, continuous dataset. This approach is efficient and scalable, making it ideal for handling large amounts of data.
```

```python
# Get the API 

r = requests.get('https://www.ncdc.noaa.gov/cdo-web/api/v2/data?datasetid=GHCND&datatypeid=TAVG&limit=1000&stationid=GHCND:USW00014732&startdate=2022-01-01&enddate=2022-12-31', headers={'token':Token})
```

```python
# Load the api response as a json

d = json.loads(r.text)
```

```python
d
```

```python
# Secure all items in the response that correspond to TAVG

avg_temps = [item for item in d['results'] if item['datatype']=='TAVG']
```

```python
# Get only the date field from all average temperature readings

dates_temp = [item['date'] for item in avg_temps]
```

```python
# Get the temperature from all average temperature readings

temps = [item['value'] for item in avg_temps]
```

```python
temps
```

```python
# Put the results in a dataframe

df_temp = pd.DataFrame()
```

```python
# Get only date and cast it to date time; convert temperature from tenths of Celsius to normal Celsius

df_temp['date'] = [datetime.strptime(d, "%Y-%m-%dT%H:%M:%S") for d in dates_temp]
df_temp['avgTemp'] = [float(v)/10.0 for v in temps]
```
