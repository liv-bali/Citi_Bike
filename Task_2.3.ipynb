## Aggregate trips

```python
df.columns
```

```python
# Groupby the count of the trips

df_group= pd.DataFrame(df.groupby(['date'])['ride_id'].count()).reset_index()
```

```python
df_group.head()
```

```python
df_group.tail()
```

```python
df_group.rename(columns = {'ride_id':'bike_rides_daily'}, inplace = True)
df = df.merge(df_group, on = "date", how = 'outer', indicator = True)
print(df['_merge'].value_counts(dropna = False))
print("Shape of January 1st is", df[df['date'] == '2018-01-01'].shape) # Check 
print("Shape of January 2nd is", df[df['date'] == '2018-01-02'].shape) # Second check 

df_temp = df.set_index('date')

print(df_temp.columns)
```

```python
df_temp.head()
```

```python
# Convert 'date' index to datetime format if not already
df_temp.index = pd.to_datetime(df_temp.index)

# Filter out the year 2021
df_filtered = df_temp[(df_temp.index < '2021-01-01') | (df_temp.index > '2021-12-31')]

# Plot the filtered data
plt.figure(figsize=(10, 5))
plt.title('Temperature per trips')
plt.plot(df_filtered['avgTemp'])
plt.plot(df_filtered['bike_rides_daily'], color='red')
plt.show()
```

## OO approach

```python
# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot bike rides
ax.plot(df_filtered['bike_rides_daily'], color='red', label='Bike Rides')

# Add labels and title
ax.set_title('Bike Rides Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Number of Rides')
ax.legend()

# Show plot
plt.show()
```

```python
## 1. **Bike Rides Over Time**  
# Created a `Figure` and `Axes` object using `plt.subplots()`.  
# Plotted the `bike_rides_daily` data on the primary y-axis.  
# Added labels, a title, and a legend for clarity.  
```

```python
# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot bike rides and temperature
ax.plot(df_filtered['bike_rides_daily'], color='red', label='Bike Rides')
ax.plot(df_filtered['avgTemp'], color='blue', label='Average Temperature')

# Add labels, title, and legend
ax.set_title('Bike Rides and Temperature Over Time')
ax.set_xlabel('Date')
ax.set_ylabel('Values')
ax.legend()

# Show plot
plt.show()
```

```python
## 2. **Bike Rides and Temperature on the Same Y-Axis**  
# Plotted both `bike_rides_daily` and `avgTemp` on the same y-axis.  
# Used different colors (`red` for bike rides and `blue` for temperature) to distinguish between the two datasets.  
# Added appropriate labels and a legend.  
```

```python
# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot bike rides on the primary y-axis
ax.plot(df_filtered['bike_rides_daily'], color='red', label='Bike Rides')
ax.set_ylabel('Bike Rides', color='red')

# Create a twin y-axis for temperature
ax2 = ax.twinx()
ax2.plot(df_filtered["avgTemp"], color="blue", label='Average Temp')
ax2.set_ylabel('Average Temperature', color='blue')

# Set title and labels
ax.set_title('Bike Rides and Temperature Over Time')
ax.set_xlabel('Date')

# Show plot
plt.show()
```

```python
## 3. **Bike Rides and Temperature on a Dual Y-Axis**  
# Used `ax.twinx()` to create a secondary y-axis for temperature.  
# Plotted `bike_rides_daily` on the primary y-axis in red.  
# Plotted `avgTemp` on the secondary y-axis in blue.  
# Labeled both axes to ensure clarity. 
```

```python
### **Paradigm Used**  
# I chose the **Object-Oriented (OO) interface** in Matplotlib because:  
# It provides better control over multiple subplots and dual y-axes.  
# It allows for more customization and flexibility.  
# It is the preferred approach for complex visualizations.  

# By structuring the plots this way, I ensured a clean and professional visualization of bike rides and temperature trends over time.
```

## Final dual plot

```python
# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 5))

# Plot bike rides on the primary y-axis
ax.plot(df_filtered['bike_rides_daily'], color="navy")
ax.set_xlabel("Year 2022", fontsize=14)  # Set x-axis label
ax.set_ylabel("Bike rides daily", color="navy", fontsize=14)  # Set y-axis label

# Create a twin y-axis for temperature
ax2 = ax.twinx()
ax2.plot(df_filtered["avgTemp"], color="red")
ax2.set_ylabel("Average temperatures", color="red", fontsize=14)

# Set title
plt.title('Temperature and Trips Over Time', fontsize=18)

# Show plot
plt.show()
```

```python
plt.style.use('ggplot')
```

```python
# Create figure and axis objects
fig, ax = plt.subplots(2, figsize=(8, 6))

# Plot bike rides on the first subplot
ax[0].plot(df_filtered['bike_rides_daily'], color="navy")
ax[0].set_ylabel("Bike rides daily", color="navy", fontsize=14)  # Set y-axis label

# Plot temperature on the second subplot
ax[1].plot(df_filtered["avgTemp"], color="red")
ax[1].set_ylabel("Average temperatures", color="red", fontsize=14)

# Set overall title
plt.suptitle('Temperature and Trips Over Time', fontsize=18)

# Show plot
plt.show()
```
